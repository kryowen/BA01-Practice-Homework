```python
w = torch.rand(1, requires_grad=True)
b = torch.rand(1, requires_grad=True)

print(w , b)
print(w.requires_grad, b.requires_grad)
print(w.grad, b.grad)
```

```python
x = torch.rand(1024)
y = 2.0 * torch.FloatTensor(x) + 5.0
w = torch.rand(1, requires_grad=True)
b = torch.rand(1, requires_grad=True)
print(w, b)
print(w.grad, b.grad)
```

```python
from torch.optim import SGD
opti = SGD([w, b], lr=1e-2)

for i in range(1000):
    h = w * x + b
    err = torch.mean((h - y) ** 2)

    opti.zero_grad() # Gradient 초기화
    err.backward() # Back Propagation
    opti.step() # Gradient Update

    if (i % 50) == 0:
        print(w, b)
```
