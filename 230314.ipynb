```python
import numpy as np
import matplotlib.pyplot as plt

w_init = np.random.rand()
# 이전에는 학습 테크닉의 부족으로 초기 값을 랜덤으로 부여하면 학습이 안됨. 학습 전에 초기값을 학습하는 부분이 필요 >> RBM 사용
# 시간이 되면 Glorot, HM 방식 소개
print(w_init)
```

```python
alpha = 0.1
epoch = 50
```

```python
w = w_init
w_list = []

for i in range(epoch):
    w_list.append(w)
    w = w - alpha * w

w_list = np.array(w_list)
```

```python
xmin = -1.0
xmax = 1.0
x = np.linspace(xmin, xmax, 100)
y = x ** 2

plt.plot(x, y, linestyle='dashed', color='b')
plt.plot(w_list, w_list**2, linestyle='None', marker='o', color='r')
plt.show()
```
